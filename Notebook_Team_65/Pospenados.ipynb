```python
# Import the Pandas package
import pandas as pd

# Import the NumPy package
import numpy as np
```

```python
# Load a file as a DataFrame and assign to df
df = pd.read_csv("Data/Reincidencia01062020.csv", sep = ',', engine = 'python', encoding = 'utf-8')
ubicacion = pd.read_csv("Data/Ubicacion_centros.csv")
```

```python
# Sizing the Data Base
num_row, num_col = df.shape
print("Database Info --->")
print("The number of records is: "+ str(num_row))
print("The number of features is: "+ str(num_col))
```

```python
# Include Longitude and Latitude in the original dataframe
df = pd.merge(df,  
                     ubicacion,  
                     on ='ESTABLECIMIENTO',  
                     how ='left')
```

```python
df.head()
```

### Format Standarization
<br> In order to standarize the database, all the capital letters will be lowered (Titles and Records)

```python
# Redefine column names
df.columns = [
    "interno",
    "situacion_juridica",
    "delito",
    "fecha_captura",
    "fecha_ingreso",
    "año_ingreso",
    "mes_ingreso",
    "estado_ingreso",
    "edad",
    "genero",
    "reincidencias",
    "establecimiento",
    "regional",
    "lon",
    "lat",
]
```

```python
# String capitalization
df['delito'] = df['delito'].apply(str.capitalize)
df['genero'] = df['genero'].apply(str.capitalize)
```

```python
# Convert Dates into datetime64 format
df['fecha_captura'] = pd.to_datetime(df['fecha_captura'], format='%d/%m/%Y')
df['fecha_ingreso'] = pd.to_datetime(df['fecha_ingreso'], format='%d/%m/%Y')
```

## Gender Analysis
Gender analysis (Unique ID)

```python
total_genero = df2.groupby(by = 'genero', as_index = False).agg({'interno': pd.Series.nunique})
total = total_genero['interno'][0] + total_genero['interno'][1]
total_genero['%'] = np.round(total_genero['interno'] / total * 100, decimals=1)
total_genero.sort_values(['interno'], ascending = False)
```

## Reoffend Analysis
How many times a prisoner have reoffended? (Unique ID)

```python
presos = df2['interno'].value_counts()
presos = presos.to_frame()
# Redefine column names
presos.columns = ["total_reincidentes"]
presos = presos['total_reincidentes'].value_counts()
```

```python
reincidencias = presos.to_frame()
reincidencias = reincidencias.sort_index(axis = 0)
```

```python
reincidencias = reincidencias.reset_index()
reincidencias.columns = ['num_reincidencias', 'total_reincidentes']
reincidencias['%'] = np.round(reincidencias['total_reincidentes'] / total * 100, decimals=3)
```

# Map

```python
centros = df2['establecimiento'].value_counts()
centros = centros.to_frame()
# Redefine column names
centros.columns = ["total_reincidentes_por_centro"]
centros = centros.reset_index()
centros.columns = ['ESTABLECIMIENTO', 'total_reincidentes_por_centro']
centros = pd.merge(centros,  
                     ubicacion,  
                     on ='ESTABLECIMIENTO',  
                     how ='left')
centros.columns = ['establecimiento', 'total_reincidentes_por_centro', 'lon', 'lat']
```

```python
import plotly.express as px
fig = px.density_mapbox(centros, lat='lat', lon='lon', z='total_reincidentes_por_centro', radius=30,
                        center=dict(lat=0, lon=180), zoom=0,
                        mapbox_style="stamen-terrain")
fig.show()
```
